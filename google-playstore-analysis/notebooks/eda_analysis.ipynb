Using pandas we cleaned this original dataset. This helps us removing duplicate, finding null values and giving appropriate data type to each column. After cleaning we did some basic analysis on our dataset.

```python
df = pd.read_excel("/content/drive/MyDrive/EDA/EDA/GooglePlayStore EDA/cleaned_dataset.xlsx") # Dataset Loading(If running in Google Colab, change the CSV path accordingly after mounting Google Drive.)
```

##Total number of apps in each category

```python
df['Category'].value_counts().head(10).plot(kind='barh')
```

##Plot showing the comparison between different Types and different Content Ratings


```python
plt.figure(figsize = (10, 8))
category = ["Type", "Content Rating"]
for i in range(0, len(category)):
    plt.subplot(2, 2, i+1)
    sns.countplot(x = df[category[i]], palette = "Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation = 45)
    plt.tight_layout()
```

##Number of Installs in each Category

```python
grouped_df = df.groupby(['Category', 'Type'])['Installs'].sum().sort_values(ascending = False).reset_index()
plt.figure(figsize=(17, 6))
sns.barplot(data=grouped_df, x='Category', y='Installs', hue='Type')
plt.title('Total Installs by Category')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

##Number of Installs in Top 10 Categories

```python
top10_df = df.groupby(['Category', 'Type'])['Installs'].sum().sort_values(ascending = False).reset_index().head(10)
plt.figure(figsize=(20, 6))
sns.barplot(data=top10_df, x='Category', y='Installs', hue='Type')
plt.title('Total Installs by top 10 Category')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

##Number of Intalls in each Type

```python
type_installs =df.groupby('Type')['Installs'].sum().sort_values(ascending = False).reset_index()
type_installs
```

##Plot showing the number of installs of Paid apps vs Free apps.


```python
plt.figure(figsize=(5, 4))
sns.barplot(data=type_installs, x='Type', y='Installs')
plt.title('Total Installs by Category and Type')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

##Top 10 categories with rating above 4


```python
df_rating = df[df['Rating']>4]
top_categories = df_rating['Category'].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(
    x=top_categories.head(10).values,
    y=top_categories.head(10).index,
    palette='crest'
)
plt.title('Top 10 Categories with Rating > 4')
plt.xlabel('Number of Apps')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

##Average Rating per Category (For apps with rating above 4)

```python
avg_rating_per_category = df_rating.groupby('Category')['Rating'].mean().sort_values(ascending=False)
plt.figure(figsize=(15, 6))
sns.barplot(
    x=avg_rating_per_category.values,
    y=avg_rating_per_category.index,
    palette='viridis'
)
plt.title('📊 Average Rating per Category (For Apps with Rating > 4)', fontsize=14)
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.xlim(4, 5)  # Since we're only looking at ratings > 4
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
