```python
def extract_max_version(x):
    if pd.isna(x):
        return np.nan
    x = str(x)
    if 'Varies' in x:
        return np.nan
    if '-' in x:
        max_version = x.split('-')[1].strip()
    elif 'W' in x:
        max_version = x.replace("W", "").strip()
    elif 'and up' in x:
        max_version = x.split()[0].strip()
    else:
        max_version = x.strip().split()[0]
    parts = max_version.split('.')
    if len(parts) >= 2:
        return f"{parts[0]}.{parts[1]}"
    return max_version
```

```python
df_copy['Android Ver'] = df_copy['Android Ver'].apply(extract_max_version)
```

```python
INVALID_VALUES = {
    "v", "Cannonau", "Varies", "Pu.0", "PN.1", "MyMetro.0", "a.2", "H7SuV0.1", "acremotecontro.0",
    "airConditionerMaintenance.0", "c.0", "Version.0", "Z.0", "Cow.0", "Fina.0", "Arrow.0",
    "no.0", "fcd822.0", "he.0", "CL.0", "owser4Craigs.0", "A.0", "Coyright.0", "Gratis.0",
    "Y4W.0", "KM.0", "INDIIMO.0", "rans.0", "HPs.0", "DPSAUS1.0", "Human.0", "Paid.0", "Pro.0", "F2i.0"
}

def curr_version(x):
    if pd.isna(x):
        return np.nan

    x = str(x).strip()

    if not x or any(invalid in x for invalid in INVALID_VALUES):
        return np.nan

    x = x.replace('Version:', '')
    x = x.replace(',', '.').replace(';', '.')
    x = x.replace('|', '')
    x = x.replace('T', '')
    x = x.replace('P', '')
    x = x.replace('p', '')
    x = x.replace('V', '')
    x = x.replace('iu', '')
    x = x.replace('eie', '')

    for delimiter in ['-', ' ', '_', '(', 'R', 'B', 'b', 'eta', 'gg', 'f', 'g', 'a']:
        if delimiter in x:
            x = x.split(delimiter)[0].strip()
            break

    x = ''.join(char for char in x if char.isdigit() or char == '.')

    parts = x.split('.')
    if len(parts) >= 2:
        return f"{parts[0]}.{parts[1]}"
    elif len(parts) == 1 and parts[0].isdigit():
        return f"{parts[0]}.0"
    else:
        return np.nan
```

```python
df_copy['clean Ver'] = df_copy['Current Ver'].apply(curr_version)
```
